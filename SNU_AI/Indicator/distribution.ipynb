```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = [
    '/home/sky3alfory/NCIA_ESS/Indicator/normal_values.parquet',
    '/home/sky3alfory/NCIA_ESS/Indicator/overcharge_values.parquet',
    '/home/sky3alfory/NCIA_ESS/Indicator/overdischarge_values.parquet'
]
```

```python

normal_df = pd.read_parquet(data[0])
overcharge_df = pd.read_parquet(data[1])
overdischarge_df = pd.read_parquet(data[2])

# SOC_slope 값 리스트로 저장
slope_values = [
    normal_df['SOC_slope'],
    overcharge_df['SOC_slope'],
    overdischarge_df['SOC_slope']
]

# Scatter plot 그리기
plt.figure(figsize=(10, 6))

# 각 타입에 대해 scatter plot을 그립니다.
for i, values in enumerate(slope_values):
    x = np.random.normal(i + 1, 0.04, size=len(values))  # 약간의 jitter 추가
    plt.scatter(x, values, alpha=0.6, label=['Normal', 'Overcharge', 'Overdischarge'][i])

plt.xticks([1, 2, 3], ['Normal', 'Overcharge', 'Overdischarge'])
plt.xlabel('Battery Type')
plt.ylabel('SOC Slope')
plt.title('Distribution of SOC Slope by Battery Type')
plt.legend()
plt.show()
```
